```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Parameters
num_users = 100
user_types = ['patient', 'doctor']
devices = ['mobile', 'web']

# Initialize list
data = []

for i in range(num_users):
    user = f"user_{i+1}"
    user_type = random.choice(user_types)
    device = random.choice(devices)
    start_time = datetime.now() - timedelta(days=random.randint(0,30))
    
    # Event sequence
    data.append([user, 'app_open', user_type, device, start_time])
    if random.random() < 0.9:  # 90% start signup
        signup_start_time = start_time + timedelta(minutes=random.randint(0,5))
        data.append([user, 'signup_start', user_type, device, signup_start_time])
        
        if random.random() < 0.6:  # 60% complete signup
            signup_complete_time = signup_start_time + timedelta(minutes=random.randint(1,5))
            data.append([user, 'signup_complete', user_type, device, signup_complete_time])
            
            if random.random() < 0.7:  # 70% book appointment
                book_time = signup_complete_time + timedelta(minutes=random.randint(1,10))
                data.append([user, 'book_appointment', user_type, device, book_time])
                
                if random.random() < 0.1:  # 10% cancel
                    cancel_time = book_time + timedelta(minutes=random.randint(1,10))
                    data.append([user, 'cancel_appointment', user_type, device, cancel_time])
            
            if random.random() < 0.5:  # 50% submit feedback
                feedback_time = signup_complete_time + timedelta(minutes=random.randint(5,15))
                data.append([user, 'submit_feedback', user_type, device, feedback_time])

# Create DataFrame
df = pd.DataFrame(data, columns=['user_id','event_name','user_type','device','timestamp'])

# Save to CSV
df.to_csv('ux_telemetry_simulated_data.csv', index=False)
print("Simulated data saved to ux_telemetry_simulated_data.csv")
```
